```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
eim = 0.001
eps = 0.00001
Nm = 30
phi_values = np.linspace(-1, 1, 500)

# Function to construct the matrix
def construct_matrix(Nm, phi, eps, theta):
    tf = np.exp(+1j * theta)
    tb = np.exp(-1j * theta)
    
    Nsys = 7 * Nm + 1
    mat = np.eye(Nsys, dtype=complex) * eps

    for i in range(1, Nm):
        mat[7*i - 6 , 7*i - 5] = tb
        mat[7*i - 5, 7*i - 6] = tf

        mat[7*i - 5, 7*i - 3] = tb
        mat[7*i - 3, 7*i - 5] = tf

        mat[7*i - 3, 7*i - 1] = tb
        mat[7*i - 1, 7*i - 3] = tf

        mat[7*i - 1, 7*i + 1] = tb
        mat[7*i + 1 ,7*i - 1] = tf

        mat[7*i - 6 , 7*i - 4] = tf
        mat[7*i  - 4,  7*i - 6] = tb

        mat[7*i - 4, 7*i - 2] = tf
        mat[7*i - 2, 7*i - 4] = tb

        mat[7*i - 2, 7*i] = tf
        mat[7*i , 7*i - 2] = tb

        mat[7*i, 7*i + 1] = tf
        mat[7*i + 1, 7*i] = tb
        
    return mat

# Store eigenvalues
eigenvalues_vs_phi = []

for phi in phi_values:
    theta = (np.pi * phi) / 4
    mat = construct_matrix(Nm, phi, eps, theta)
    eigenvalues = np.linalg.eigvals(mat)
    eigenvalues_vs_phi.append(eigenvalues)

# Convert to numpy array for easier manipulation
eigenvalues_vs_phi = np.array(eigenvalues_vs_phi, dtype=object)

# Remove flat band states at E = 0, +-0.5
filtered_eigenvalues_vs_phi = []
for i in range(eigenvalues_vs_phi.shape[0]):
    filtered = eigenvalues_vs_phi[i][np.abs(eigenvalues_vs_phi[i].real) > 1e-3]  # Remove E=0
    filtered_eigenvalues_vs_phi.append(filtered)

# Calculate slopes at the Dirac point
dirac_index = np.argmin(np.abs(phi_values))
slopes = []

for j in range(len(filtered_eigenvalues_vs_phi[dirac_index])):
    try:
        e_left = filtered_eigenvalues_vs_phi[dirac_index - 1][j].real
        e_right = filtered_eigenvalues_vs_phi[dirac_index + 1][j].real
        slope = (e_right - e_left) / (phi_values[dirac_index + 1] - phi_values[dirac_index - 1])
        slopes.append(slope)
    except IndexError:
        continue
    except ValueError:
        # If complex eigenvalues are encountered, skip them
        continue

# Ensure filtered eigenvalues are real numbers
energy_levels = np.array([ev.real for ev in filtered_eigenvalues_vs_phi[dirac_index]], dtype=float)

# Plot the slope vs. energy levels at Dirac point
plt.figure(figsize=(10, 6))
plt.plot(energy_levels, slopes, 'bo', markersize=5)
plt.xlabel('Energy at Dirac Point (E)', fontname='Times New Roman', fontsize=29, fontweight='normal')
plt.ylabel('Slope dE/dΦ', fontname='Times New Roman', fontsize=29, fontweight='normal')

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.ylim(-2.5,2.5)
#plt.xlim(-0.2,0.2)
plt.grid(True)
plt.show()

# Print the slopes
# Find the slope corresponding to the energy closest to E=0
closest_index = np.argmin(np.abs(energy_levels))
slope_at_zero_energy = slopes[closest_index]

# Print the slope at E = 0
print("Slope at E = 0:", slope_at_zero_energy)
```
