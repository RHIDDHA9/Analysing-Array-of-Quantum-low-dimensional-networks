```python
#Dual Diamond with N=2

import numpy as np
import matplotlib.pyplot as plt


eim = 0.001
eps = 0.00001
Nm = 30
Nsys = 22 * Nm + 2
t =0.5
phi = 0.5
theta = (np.pi * phi) / 6
tf = 1 * np.exp(+1j * theta)
tb = 1 * np.exp(-1j * theta)

# Construct the Hamiltonian matrix
mat = np.eye(Nsys, dtype=complex ) * eps

for i in range(1, Nm):
        mat[22*i - 20, 22*i - 17] = (tb)
        mat[22*i - 17, 22*i - 20] = (tf)
        
        mat[22*i - 17, 22*i - 13] = (tb)
        mat[22*i - 13, 22*i - 17] = (tf)
        
        mat[22*i - 13, 22*i - 9] = (tb)
        mat[22*i - 9, 22*i - 13] = (tf)
        
        mat[22*i - 9, 22*i - 5] = (tb)
        mat[22*i - 5 ,22*i - 9] = (tf)
        
        mat[22*i - 5 , 22*i - 1] = (tb)
        mat[22*i - 1,  22*i - 5] = (tf)
        
        mat[22*i - 1, 22*i + 2] = (tb)
        mat[22*i + 2, 22*i - 1] = (tf)
        
        mat[22*i - 20, 22*i - 16] = (tf)
        mat[22*i - 16, 22*i - 20] = (tb)
        
        mat[22*i - 16, 22*i - 12] = (tf)
        mat[22*i - 12, 22*i - 16] = (tb)

        mat[22*i - 12 , 22*i - 8] = (tf)
        mat[22*i - 8,  22*i - 12] = (tb)
        
        mat[22*i - 8, 22*i - 4] = (tf)
        mat[22*i - 4, 22*i - 8] = (tb)
        
        mat[22*i - 4, 22*i] = (tf)
        mat[22*i , 22*i - 4] = (tb)
        
        mat[22*i , 22*i + 2] = (tf)
        mat[22*i + 2, 22*i] = (tb)

        mat[22*i - 21, 22*i - 20] = t
        mat[22*i - 20, 22*i - 21] = t
        
        mat[22*i - 21, 22*i - 19] = (tb)
        mat[22*i - 19, 22*i - 21] = (tf)
        
        mat[22*i - 19, 22*i - 15] = (tb)
        mat[22*i - 15, 22*i - 19] = (tf)
        
        mat[22*i - 15, 22*i - 11] = (tb)
        mat[22*i - 11 ,22*i - 15] = (tf)
        
        mat[22*i - 11 , 22*i - 7] = (tb)
        mat[22*i - 7,  22*i - 11] = (tf)
        
        mat[22*i - 7, 22*i - 3] = (tb)
        mat[22*i - 3, 22*i - 7] = (tf)
        
        mat[22*i - 3, 22*i + 1] = (tb)
        mat[22*i + 1, 22*i - 3] = (tf)
        
        mat[22*i + 1, 22*i - 2] = (tb)
        mat[22*i - 2, 22*i + 1] = (tf)

        mat[22*i - 2 , 22*i - 6] = (tb)
        mat[22*i - 6,  22*i - 2] = (tf)
        
        mat[22*i - 6, 22*i - 10] = (tb)
        mat[22*i - 10, 22*i - 6] = (tf)
        
        mat[22*i - 10, 22*i - 14] = (tb)
        mat[22*i - 14, 22*i - 10] = (tf)
        
        mat[22*i - 14, 22*i - 18] = (tb)
        mat[22*i - 18, 22*i - 14] = (tf)

        mat[22*i - 18, 22*i - 21] = (tb)
        mat[22*i - 21, 22*i - 18] = (tf)

        mat[22*i + 1, 22*i + 2] = (t)
        mat[22*i + 2, 22*i + 1] = (t)

        

# Diagonalize the Hamiltonian matrix to find eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(mat)


energy_I_list = []

# Calculate I for each wave function (eigenvector)
for idx in range(len(eigenvalues)):
    wave_function = eigenvectors[:, idx]
    #wave_function_squared = (wave_function)**2
    sum_psi_squared = np.sum(np.abs(wave_function) ** 2)  
    sum_psi_fourth = np.sum(np.abs(wave_function) ** 4)   

    I = sum_psi_fourth / (sum_psi_squared ** 2)
    energy_I_list.append([eigenvalues[idx], I])


energy_values = [pair[0] for pair in energy_I_list]
I_values = [pair[1] for pair in energy_I_list]


plt.figure(figsize=(8, 6))
plt.scatter(energy_values, I_values, color='blue', s=10)
plt.xlabel('E', fontsize=29, fontname='Times New Roman', fontweight='bold')
plt.ylabel('I', fontsize=29, fontname='Times New Roman', fontweight='bold')
plt.xticks(fontsize=20)
plt.yticks(fontsize=15)
plt.grid(True, linestyle='--', color='pink')
plt.ylim(0,1.05)
plt.xlim(-4,4)
plt.show()
```
